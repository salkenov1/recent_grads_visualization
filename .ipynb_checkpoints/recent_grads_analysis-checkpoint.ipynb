# Data Visualization Project

In this project, we'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by *American Community Survey*, which conducts surveys and aggregates the data.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

| Rank           | Rank by median earnings (the dataset is ordered by this column)            |
|----------------|----------------------------------------------------------------------------|
| Major_code     | Code of the major                                                          |
| Major          | Major description                                                          |
| Major_category | Category of major from Carnevale et al                                     |
| Total          | Total number of people with major                                          |
| Sample_size    | Sample size (unweighted) of full-time, year-round ONLY (used for earnings) |
| Men            | Male graduates                                                             |
| Women          | Female graduates                                                           |
| ShareWomen     | Women as share of total                                                    |
| Employed       | Number employed (ESR == 1 or 2)                                            |
| Median         | Median earnings of full-time, year-round workers                           |
| Low_wage_jobs  | Number in low-wage service jobs                                            |
| Full_time      | Employed 35 hours or more                                                  |
| Part_time      | Employed less than 35 hours                                                |

Using visualizations, we can start to explore questions from the dataset like:

- Do students in more popular majors make more money? (Using scatter plots)
- How many majors are predominantly male? Predominantly female? (Using histograms)
- Which category of majors have the most students? (Using bar plots)

We'll explore how to do these and more while primarily working in pandas. Before we start creating data visualizations, let's import the libraries we need and remove rows containing null values.

First of all, let's setup our environment by importing the libraries. We will need pandas, numpy and matplotlib. 

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from pandas.plotting import scatter_matrix 

%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv", sep=",", index_col=0)
first_row = recent_grads.iloc[0]
recent_grads.head(5)
```

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print("Raw data count is:", raw_data_count)
print("Cleaned data count is:", cleaned_data_count)
```

As seen from the cell above, there is only one row removed during the cleaning process with DROPNA.

```python
recent_grads.plot(x="Sample_size", y="Median", title="Fig 1. Sample size vs Median", kind="scatter", figsize=(5,5))
```

According to the graph above, the students of less popular majors are tend to get salary between 20000\\$ and 60000\\$, there are less grads who get between 60000\\$ and 75000\\$, and there is some, major students of which get more than 100000\\$. The students of more popular majors, where the number of grads is between 2000 and 3000, are tend to be paid between 30000\\$ and 50000\\$. The big flow of recent grads coming from the most popular major resulted in oversupply on the labour market and stagnated their salary around 40000\\$.

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", title="Fig 2. Sample size vs Unemployment rate", kind="scatter", figsize=(5,5))
```

```python
recent_grads["Unemployment_rate"].describe()
```

It is clear in the graph above that the majority of unemployed people (above 0.100 unemployment rate) come from less popular major. In contrast, the lowest unmployment rate is also possessed by less popular majors. It clearly indicates the importance of the major itself, not the number of students pursuing it. However, the most popular major, with the highest number of graduates, shows relatively better results in terms of employment, even though the rate is higher than the mean (0.068).

```python
median_std = np.std(recent_grads["Median"])
median_mean = np.mean(recent_grads["Median"])
upper_level = median_mean + median_std
lower_level = median_mean - median_std
print("The upper level of median is:", upper_level)
print("The lower level of median is:", lower_level)
```

- **Do students in more popular majors make more money?**

According to the graph above (Fig 1.), the students of less popular majors are tend to get salary between 20,000\\$ and 60,000\\$, there are less grads who get between 60,000\\$ and 76,000\\$, and there is some, major students of which get more than 100,000\\$. The students of more popular majors, where the number of grads is between 2,000 and 3,000, are tend to be paid between 30,000\\$ and 50,000\\$. The big flow of recent grads coming from the most popular major resulted in oversupply on the labour market and stagnated their salary around 40,000\\$.

- **Do students that majored in subjects that were majority female make more money?**

According to Fig. 7., the grads of women-free majors are likely to get salary around 40,000\\$, which is close to the average salary. The median salary of grads, where the share of women is between 0.2 and 0.5, shows diversity in terms of it's range. The grads get between 25,000\\$ and 78,000\\$. However, there is also a major with around 0.16 women rate, that's shows an outlier behaviour. The grads of this major get more than 100,000\\$, which is twice bigger then the upper level. The students coming from women-occupied majors show a tendency of getting the lower level of the median salary or even below, as well as not getting higher than the upper level. 

- **Is there any link between the number of full-time employees and median salary?**

As shown in Fig. 3., the median salary of less occupied full-time jobs (less than 50,000), varies in broad spectrum between 22,000\\$ and 76,000\\$. However, there is a major, full-time employees of which get more than 100,000\\$. As the number of full-time employees grows, the median salary range shrinks accordingly. So the median salary range between 38,000\\$ and 60,000\\$ narrows to the range between 39,000\\$ and 43,000\\$ as it reaches the number of 175,000 full-time employed people. The median salary of the most occupied job is around 38,000\\$. According to this tendency, the chances of getting higher than the average or even the upper-level of salary are much bigger in less employed positions. As the number of employees increases, the chance of getting the average salary increases accordingly.

```python
fig, ax = plt.pyplot.subplots()
ax = recent_grads["Sample_size"].hist(bins=20)
ax.set_title("Fig 8. Distribution of Sample size of full-time grads ")
ax.set_ylabel("# of jobs")
ax.set_xlabel("# of full-time employees")
```

```python
fig, ax = plt.pyplot.subplots()
ax = recent_grads["Median"].hist()
ax.set_title("Fig 9. Distribution of Median salary")
ax.set_ylabel("# of grads having it")
ax.set_xlabel("Median salary")
```

```python
fig, ax = plt.pyplot.subplots()
ax = recent_grads["ShareWomen"].hist()
ax.set_title("Fig 10. Distribution of Women grads from majors")
ax.set_ylabel("# of majors")
ax.set_xlabel("% of women")
```

```python
fig, axs = plt.pyplot.subplots(1, 2, sharey=True, figsize=(15.0, 6.0))


axs[0].hist(recent_grads["Men"])
axs[0].set_title("Fig 11. Male graduates")
axs[0].set_xlabel("# of grads")
axs[0].set_ylabel("# of majors")
axs[1].hist(recent_grads["Women"])
axs[1].set_title("Fig 12. Female graduates")
axs[1].set_xlabel("# of grads")
axs[1].set_ylabel("# of majors")
```

- **What percent of majors are predominantly male? Predominantly female?**

The majors are mostly dominated by male. The percentage of majors dominated by women is around ~20% (ShareWomen > 0.6). It means the rest 80% of majors are significantly hegemonized by male.


- **What's the most common median salary range?**

The most common median salary varies between 25,000\\$ and 39,500\\$

```python
fig, axs = plt.pyplot.subplots(1, 2, figsize = (15, 7), sharey = True)

axs[0].bar(np.arange(10), recent_grads[:10]["ShareWomen"], width = 0.65)
axs[1].bar(np.arange(10), recent_grads[-10:]["ShareWomen"], width = 0.65)

axs[0].set_title("Fig 13. Percentage of women, first 10")
axs[0].set_ylabel("Women as share")
axs[0].set_xlabel("# of entries")

axs[1].set_title("Fig 14. Percentage of women, last 10")
axs[1].set_ylabel("Women as share")
axs[1].set_xlabel("# of entries")
```

As shown in Figures 13 and 14, the first 10 rows show that women' share is significantly lower than in the last 10 rows.

```python
fig, axs = plt.pyplot.subplots(1, 2, figsize = (12, 7), sharey = True)

axs[0].bar(np.arange(10), recent_grads[:10]["Unemployment_rate"], width = 0.65)
axs[1].bar(np.arange(10), recent_grads[-10:]["Unemployment_rate"], width = 0.65)

axs[0].set_title("Fig 15. Unemployment rate, first 10")
axs[0].set_ylabel("Rate")
axs[0].set_xlabel("# of entries")

axs[1].set_title("Fig 16. Unemployment rate, last 10")
axs[1].set_ylabel("Rate")
axs[1].set_xlabel("# of entries")
```
